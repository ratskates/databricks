```python
from pyspark.sql import SparkSession
# Create SparkSession 
spark = SparkSession.builder \
      .master("local[8]") \
      .appName("SparkByExamples.com") \
      .getOrCreate() 


df = spark \
.read \
.format("csv") \
.option("inferSchema","true") \
.option("header","true") \
.load("/home/garry/development/da_data_repo/hotels-vienna/raw/hotelbookingdata-vienna.csv") \

df.select("hotel_id").orderBy("hotel_id",ascending=True).show(100)

df.printSchema()


df2 = spark.read.option("header",True) \
          .csv("/home/garry/development/da_data_repo/hotels-vienna/raw/hotelbookingdata-vienna.csv") \
          .createOrReplaceTempView("HOTELS")


spark.sql("SELECT hotel_id FROM HOTELS") \
     .show(5)

    
```
